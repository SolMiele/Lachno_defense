```python
import os
import glob
import gzip
from Bio import SeqIO
```

```python
# Set the root directory where compressed GenBank files are stored
root_dir = "Z:/Projets/projet_DKA/Solange/DefenseSystems/lachno_tree/Genomes"  # Change this to your folder path
output_dir = "Z:/Projets/projet_DKA/Solange/DefenseSystems/lachno_tree/faa"  # Change this to where you want to save FASTA files
```

```python
# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Find all .gbff.gz files in subdirectories
gbff_files = glob.glob(os.path.join(root_dir, "**", "*.gbff.gz"), recursive=True)
```

```python
for gbff_filename in gbff_files:
    # Define the output .faa filename
    faa_filename = os.path.join(output_dir, os.path.basename(gbff_filename).replace(".gbff.gz", ".faa"))
    
    # Open the compressed file
    with gzip.open(gbff_filename, "rt") as input_handle, open(faa_filename, "w") as output_handle:
        for seq_record in SeqIO.parse(input_handle, "genbank"):
            print(f"Processing {gbff_filename} - GenBank record {seq_record.id}")
            for seq_feature in seq_record.features:
                if seq_feature.type == "CDS" and "translation" in seq_feature.qualifiers:
                    locus_tag = seq_feature.qualifiers.get('locus_tag', ['unknown'])[0]
                    protein_seq = seq_feature.qualifiers['translation'][0]
                    output_handle.write(f">{locus_tag} from {seq_record.name}\n{protein_seq}\n")

print("Conversion completed!")
```
